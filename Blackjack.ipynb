```python
import random
from IPython.display import clear_output
name_of_card=('Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King')
colours=('Hearts','Diamonds','Spades','Clubs')
value={'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing=True
total_chips=100
bet_value=0

class Card:
    def __init__(self,name_of_card,colour):
        self.name_of_card=name_of_card
        self.colour=colour
        
    def __str__(self):
        return ('{} of {}'.format(self.name_of_card, self.colour))
              
class Deck:
    def __init__(self):
        self.deck=[]
        
        for card in name_of_card:
            for color in colours:
                self.deck.append(Card(card,color))
                
    def __str__(self):
        deck_of_card=''
        for card in self.deck:
            deck_of_card+='\t'+card.__str__()        
        return deck_of_card
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop()
    
class Hand:
    def __init__(self):
        self.card=[]
        self.value=0
        self.aces=0
        
    def add_card(self,card):
        self.card.append(card)
        self.value+=value[card.name_of_card]
        if card.name_of_card=='Ace':
            self.aces+=1
            
    def ace_adjustment(self):
        while self.value>21 and self.aces:
            self.value-=10
            self.aces-=1
            
def win_bet(bet_value):
    global total_chips
    total_chips+=bet_value
    
def lose_bet(bet_value):
    global total_chips
    total_chips-=bet_value
    
def take_bet():
    global bet_value
    while True:
        try:
            bet_value = int(input('How many chips would you like to bet? '))
        except:
            print('Bet value must be an integer')
            
        else:
            if bet_value>total_chips:
                print("Sorry, your bet can't exceed {}".format(total_chips))
                
            else:
                break
                
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.ace_adjustment()
    
def hit_or_stand(deck,hand):
    global playing
    while True:
        x=input("Would you like to Hit or Stand? Enter 'h' or 's'")
        if x=='h':
            hit(deck,hand)
        elif x=='s':
            playing =False
            print("Player stands. Dealer is playing.")
        else:
            print("Sorry, please try again.")
            continue
        break
        

def displaycard(dealer,player):
    print("Dealer's cards\n")
    print("Card Hidden")
    print(dealer.card[1], sep='\n')
    print("*****************************")
    print("Player's card\n")
    print(*player.card, sep='\n')
    
def display_all_cards(dealer,player):
    print("Dealer's cards")
    print(*dealer.card)
    print("*****************************")
    print("Player's card")
    print(*player.card, sep='\n')
    

while True:
    deck=Deck()
    deck.shuffle()
    
    player_hand=Hand()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    dealer_hand=Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
    
    take_bet()
    
    displaycard(dealer_hand,player_hand)
    
    while playing:
        hit_or_stand(deck,player_hand)
        
        
        displaycard(dealer_hand,player_hand)
        
        if player_hand.value >21:
            lose_bet(bet_value)
            print("Player busts!")
            break
            
    #Player has stopped taking more cards, turn for dealer now
    
    if player_hand.value<=21:
        
        while dealer_hand.value<17:
            hit(deck,dealer_hand)
            
        display_all_cards(dealer_hand,player_hand)
            
        if dealer_hand.value>21:
            win_bet(bet_value)
            print("Dealer busts!")
            
        elif dealer_hand.value>player_hand.value:
            lose_bet(bet_value)
            print("Dealer wins!")
            
        elif dealer_hand.value<player_hand.value:
            win_bet(bet_value)
            print("Player wins!")
            
        else:
            print("It's a tie")
            
    print("Player chips value currnetly is ",total_chips)
    
    new_game = input("Would you like to play another hand? Enter 'y' or 'n' ")
    
    if new_game[0].lower()=='y':
        playing=True
        continue
    else:
        print("Thanks for playing!")
        break
        

    
    
        
        
```
